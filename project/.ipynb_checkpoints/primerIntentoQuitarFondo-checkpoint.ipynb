```python
def funcionLeerClase(pathDirectorio):
    pathsCompletos = glob(pathDirectorio+'/*')
    imagenes = []
    for nombreImagen in pathsCompletos:
        imagen = cv2.imread(nombreImagen)
        imagenes.append(imagen)
    return imagenes
```

```python
def funcionLeerClase2(pathDirectorio):
    pathsCompletos = glob(pathDirectorio+'/*')
    imagenes = []
    nombresMonedas = []
    for nombreImagen in pathsCompletos:
        imagen = cv2.imread(nombreImagen)
        imagenes.append(imagen)
        nombresMonedas.append(nombreImagen)
    return imagenes,nombresMonedas
```

```python
def funcionLeerTodasClases(pathDirectorio):
    etiquetasClases = ['10c','1c','1e','20c','2c','2e','50c','5c']
    lista1c = []
    lista2c = []
    lista5c = []
    lista10c = []
    lista20c = []
    lista50c = []
    lista1e = []
    lista2e = []
    vec_listas = [lista1c,lista2c,lista5c,lista10c,lista20c,lista50c,lista1e,lista2e]
    for i in range(len(etiquetasClases)):
        print('En',pathDirectorio+'/'+etiquetasClases[i], end = "")
        vec_listas[i] = funcionLeerClase(pathDirectorio+'/'+etiquetasClases[i])
        print(', hay', len(vec_listas[i]), 'ejemplos')
    return vec_listas
```

```python
def ecualizacionAdaptativa(image):
    lab = cv2.cvtColor(image, cv2.COLOR_BGR2LAB)
    lab_planes = cv2.split(lab)
    clahe = cv2.createCLAHE(clipLimit=2.0,tileGridSize=(3,3))
    lab_planes[0] = clahe.apply(lab_planes[0])
    lab = cv2.merge(lab_planes)
    ecualizada = cv2.cvtColor(lab, cv2.COLOR_LAB2BGR).astype('uint8')
    return ecualizada
```

```python
#Ya no la usamos
def imagen_media_color(imagen, nClusters):
    alg = KMeans(n_clusters=nClusters, n_init = 20)
    alg.fit(np.hstack((img[:,:,0].reshape(-1,1), img[:,:,1].reshape(-1,1), img[:,:,2].reshape(-1,1))))
    etiquetas = alg.labels_
    centros = alg.cluster_centers_
    nuevaImagen = np.zeros(imagen.shape)
    nuevaImagen = centros[etiquetas].reshape(imagen.shape)
    puntoCentro = (imagen.shape[0]//2,imagen.shape[1]//2)
    colorMoneda = nuevaImagen[puntoCentro]
    nuevaImagen = nuevaImagen/255
    return nuevaImagen, colorMoneda
```

```python
#Ya no la usamos
def quitarFondo(img):
    imgBlurreada = cv2.blur(img,(7,7))
    imagenMedia, colorMoneda = imagen_media_color(imgBlurreada, nClusters = 2)
    puntoCentro = (img.shape[0]//2,img.shape[1]//2)
    imagenSinFondo = img.copy()
    posFondo = np.any(imagenMedia != imagenMedia[puntoCentro], axis = 2)
    #Ponemos el fondo del mismo color que la moneda
    imagenSinFondo[posFondo] = colorMoneda
    #imagenSinFondo[posFondo] = 255
    return imagenSinFondo
```

```python
def realceBordes(imagen, a, tamKernel):    
    imagenConMedia = cv2.blur(imagen,(tamKernel,tamKernel))
    imageBordes = imagen - imagenConMedia
    imagenFinal = (imagen + a * imageBordes).astype(np.uint8)
    return imagenFinal
```

```python
def encontrarMonedaCentral(img, subplotCirculos = None):
    output = img.copy()
    gray = cv2.cvtColor(output, cv2.COLOR_RGB2GRAY)
    circles = cv2.HoughCircles(gray,cv2.HOUGH_GRADIENT,dp=1,minDist=150,param1 = 30,param2=50,minRadius=30,maxRadius=90)
    if circles is not None:
        circles = np.round(circles[0, :]).astype("int")
        for (x, y, r) in circles:
            #cv2.circle(output, (x, y), r, (0, 255, 0), 1)
            cv2.ellipse(output,(x, y), r,(0,255,0),1)
    if(subplotCirculos is not None):
        subplotCirculos.imshow(output)
    grayOutput = cv2.cvtColor(output, cv2.COLOR_RGB2GRAY)
    siluetaMoneda = gray-grayOutput
    siluetaMoneda[siluetaMoneda>0] = 255
    mascaraMoneda = np.copy(siluetaMoneda)
    cnts = cv2.findContours(mascaraMoneda, cv2.RETR_EXTERNAL, cv2.CHAIN_APPROX_SIMPLE)
    cnts = cnts[0] if len(cnts) == 2 else cnts[1]
    cv2.fillPoly(mascaraMoneda, cnts, (255, 255, 255))
    '''
   for c in cnts:
    cv2.drawContours(mascaraMoneda,[c], 0, (255,255,255), -1)
    '''
    img[mascaraMoneda==0] = 0
    return img
```

```python
def encontrarMonedaCentralSinQuitarFondo(img, subplotCirculos = None):
    output = img.copy()
    gray = cv2.cvtColor(output, cv2.COLOR_RGB2GRAY)
    circles = cv2.HoughCircles(gray,cv2.HOUGH_GRADIENT,1,150,param1 = 30,param2=50,minRadius=30,maxRadius=90)
    if circles is not None:
        circles = np.round(circles[0, :]).astype("int")
        for (x, y, r) in circles:
            cv2.circle(output, (x, y), r, (0, 255, 0), 4)
    if(subplotCirculos is not None):
        subplotCirculos.imshow(output)
    return output
```

```python
def umbralAdaptativo(img):    
    gray_blur = cv2.cvtColor(img, cv2.COLOR_RGB2GRAY)
    #gray_blur = cv2.GaussianBlur(gray, (15, 15), 0)
    thresholded = cv2.adaptiveThreshold(gray_blur, 255, cv2.ADAPTIVE_THRESH_GAUSSIAN_C,cv2.THRESH_BINARY_INV, 3, 1)
    return thresholded
```

```python
def cerrarHuecos(img):
    kernel = np.ones((3, 3), np.uint8)
    closing = cv2.morphologyEx(img, cv2.MORPH_CLOSE, kernel, iterations=4)
    '''
    kernel = np.ones((5, 5), np.uint8)
    closing = cv2.morphologyEx(closing, cv2.MORPH_OPEN, kernel, iterations=4)'''
    return closing
```

```python
def detectarCirculos(closing, img):
    nuevaImg = img.copy()
    cont_img = closing.copy()
    contours, hierarchy = cv2.findContours(cont_img, cv2.RETR_EXTERNAL,cv2.CHAIN_APPROX_SIMPLE)
    for cnt in contours:
        area = cv2.contourArea(cnt)
        '''
        if area < 2000 or area > 4000:
            continue'''
        if (len(cnt) < 5) or (len(cnt) > 150):
            continue
        ((x,y), r) = cv2.minEnclosingCircle(cnt)
        center = (int(x),int(y))
        cv2.circle(img = nuevaImg, center=center, radius=int(r), color=(0, 255, 0), thickness=1)
    return nuevaImg
```

```python
def aplicaPCA(imagen,nComponentes = None):
    pca = PCA(n_components=nComponentes)
    PCAs = pca.fit_transform(imagen)
    print('De media mantenemos el {}% de la información de los cuadros'.format(round(np.sum(pca.explained_variance_ratio_)*100, 2)))
    return PCAs, pca
```
